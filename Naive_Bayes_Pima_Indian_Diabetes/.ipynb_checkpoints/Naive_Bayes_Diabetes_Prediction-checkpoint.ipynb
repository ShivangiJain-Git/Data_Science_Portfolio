Several constraints were placed on the selection of these instances from a larger database. In particular, all patients here are females at least 21 years old of Pima Indian heritage.

```python
#Checking no. of outliers 
col_val = ['Preg', 'Plas', 'Pres', 'skin', 'test', 'mass', 'pedi', 'age']

for i in col_val:
    print('No. of outliers in {} is {}'.format(i, df[((df[i]-df[i].mean())/df[i].std()).abs() > 3][i].count()))
```

There are ouliers in this dataset. We can treat it. Before that we will make copy of original dataset so that original dataset remain unaffected.

```python
#Creating copy of original dataset
df1 = df.copy(deep= True)

#Lets impute outliers with Median
for i in col_val:
    Q1 = df1[i].quantile(0.25)
    Q3 = df1[i].quantile(0.75)
    IQR = Q3 - Q1
    
    low = Q1 - 1.5*IQR
    high = Q3 + 1.5*IQR
    df1.loc[(df1[i] < low) | (df1[i] > high),i] = df1[i].median()
```

```python
#Checking no. of outliers again

#Number of outliers
col_val = ['Preg', 'Plas', 'Pres', 'skin', 'test', 'mass', 'pedi', 'age']


for i in col_val:
    print('No. of outliers in {} is {}'.format(i, df1[((df1[i]-df1[i].mean())/df1[i].std()).abs() > 3][i].count()))
```

### Calculate diabetes ratio of True/False

```python
n_true = len(df1.loc[df1['class'] == True])
n_false = len(df1.loc[df1['class'] == False])
print("Number of true cases: {0} ({1:2.2f}%)".format(n_true, (n_true / (n_true + n_false)) * 100 ))
print("Number of false cases: {0} ({1:2.2f}%)".format(n_false, (n_false / (n_true + n_false)) * 100))
```

So we have 34.90% people in current data set who have diabetes and rest of 65.10% doesn't have diabetes. 

Its a good distribution True/False cases of diabetes in data.


### Splitting dataset

```python
#Independent variables
X = df1.drop(['class'], axis =1)

#Dependent variables
y = df1['class']

#Splitting the dataset in training and test set in the ratio of 70:30
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size =0.30, random_state=1)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

### Checking hidden missing values which has 0s

As we checked missing values earlier but haven't got any. But there can be lots of entries with 0 values. We must need to take care of those as well.

```python
X_train.head()
```

### Replace 0s with median/mean

```python
from sklearn.impute import SimpleImputer

si = SimpleImputer(missing_values = 0, strategy = "mean")
cols = X_train.columns

X_train = pd.DataFrame(si.fit_transform(X_train))
X_test = pd.DataFrame(si.fit_transform(X_test))

X_train.columns = cols
X_test.columns = cols
```

```python
X_train.head()
```

### Creating Model

```python
#Creating object of naive bayes

nb = GaussianNB()
nb.fit(X_train,y_train)

#Predicting on test
pred = nb.predict(X_test)

print("On Test data", nb.score(X_test,y_test))
print("On Train data", nb.score(X_train, y_train))
```

```python
#Confusion Matrix
print(confusion_matrix(y_test,pred))
```

```python
from sklearn import metrics

print("Confusion Matrix")
cm = metrics.confusion_matrix(y_test, pred, labels=[1,0])
df_cm = pd.DataFrame(cm, index = [i for i in ["1","0"]],
                    columns = [i for i in ["Predict 1", "Predict 0"]])
plt.figure(figsize=(7,5))
sns.heatmap(df_cm, annot=True)
```

The confusion matrix

True Positives (TP): we correctly predicted that they do have diabetes 60

True Negatives (TN): we correctly predicted that they don't have diabetes 124

False Positives (FP): we incorrectly predicted that they do have diabetes (a "Type I error") 22 Falsely predict positive Type I error

False Negatives (FN): we incorrectly predicted that they don't have diabetes (a "Type II error") 25 Falsely predict negative Type II error

```python
#Print classification report

print("Classification Report")
print(metrics.classification_report(y_test,pred,labels=[1,0]))
```

So, true poitive have preciosn and recall 73% and 71% respectively.

### Conclusion and Interpretation

Our model has correctly predicted with the accuracy of 79%. Overall good model.
