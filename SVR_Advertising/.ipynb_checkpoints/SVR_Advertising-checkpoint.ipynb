### Splitting dataset

```python
X = df.drop(['Sales'],axis=1)
y= df['Sales']

#Splitting of dataset in train and test in the ratio of 70:30
X_train, X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=1)
```

### Creating Model

```python
#Creating Object

reg = SVR(kernel= 'linear')
reg.fit(X,y)

pred = reg.predict(X_test)

print('On test', reg.score(X_test,y_test))
print('On train', reg.score(X_train, y_train))
print('R2 Score',r2_score(y_test,pred))
```

### Conclusion and Interpretation

So SVR model gives 93% accuracy on Test and 87% in Train
