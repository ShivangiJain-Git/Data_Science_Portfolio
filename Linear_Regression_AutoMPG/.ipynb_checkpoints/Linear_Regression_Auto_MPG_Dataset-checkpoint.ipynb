### Importing Necessary Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style='darkgrid')
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

Loading auto-mpg dataset

```python
df = pd.read_csv('auto-mpg.csv')
```

```python
#dropping 'car name' column as it is of no use
df.drop('car name', axis=1,inplace = True)
```

```python
df.head()
```

```python
#We can see origin has 3 values 1,2,3 which is not meaningful. Lets replace these with actual values
df['origin'] = df['origin'].replace({1: 'America', 2: 'Europe', 3: 'Asia'})
df.head()
```

```python
#creating dummies for column origin so that they can be evaluated while forming equation
df = pd.get_dummies(df, columns=['origin'])
df.head()
```

```python
df.shape
```

```python
#In horsepower there is data with ? we need to treat it.
df = df.replace('?',np.nan)
df.head()
```

```python
df.median()
```

```python
#Replace missing value with median
filler = lambda x : x.fillna(x.median())
df = df.apply(filler,axis=0)
df['horsepower'] = df['horsepower'].astype('float64')
```

```python
df.head()
```

Checking Outliers

```python
#Checking of ouliers through boxplot
df.boxplot(figsize=(10,5),color='green')
```

```python
#Number of outliers
col_val = df.columns
for i in col_val:
    print('No. of outliers in {} is {}'.format(i, df[((df[i]-df[i].mean())/df[i].std()).abs() > 3][i].count()))
```

### Univariate Analysis

```python
#Checking distribultion using distplot

Cols = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year']
pos = 1
plt.figure(figsize=(15,15))
for i in Cols:
    plt.subplot(3,3,pos)
    sns.distplot(df[i],hist = True, color= 'green')
    pos +=1
```

### Bivariate Analysis

```python
#chekcing correlation between independent varaibles
cor = df.corr()
plt.figure(figsize=(16,12))
sns.heatmap(cor, xticklabels=cor.columns, yticklabels= cor.columns,annot= True,cmap='magma')
```

```python
df1 = df.iloc[:,0:7]
sns.pairplot(df1,diag_kind= 'kde');
```

### Splitting the data

```python
#Independent variables
X = df.drop(['mpg','origin_Europe'], axis = 1)
#Dependent variables
y = df[['mpg']]

#splitting X and y in the ration of 70:30
X_train, X_test , y_train, y_test = train_test_split(X,y,test_size= 0.30,random_state=1)
```

### Creating model

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
for idx, cols in enumerate(X_train.columns):
    print('Coefficient of {} is {}'.format(cols, lr.coef_[0][idx]))
```

```python
#predicting on test data
pred = lr.predict(X_test)

#predicting Score
print('Score on Test data',lr.score(X_test,y_test))
print('Score on Training data',lr.score(X_train,y_train))
```

So, we got 84% accuracy score on Linear model in test data and 81% in train data. It means model is good neither overfit nor underfit.
